```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('array1_new.txt', 'r') as f:
    array1 = f.readlines()
array1 = np.array([list(map(int, row.split())) for row in array1])

with open('array2_new.txt', 'r') as f:
    array2 = f.readlines()
array2 = np.array([list(map(int, row.split())) for row in array2])

with open('result_new.txt', 'r') as f:
    result = f.readlines()
result = np.array([list(map(int, row.split())) for row in result])
```

```python
np.allclose(array1 @ array2, result)
```

```python
with open('times_openmp.txt', 'r') as f:
    data_openmp = f.readlines()
with open('times_seq.txt', 'r') as f:
    data_seq = f.readlines()
```

```python
data_openmp = np.array([list(map(int, row.split())) for row in data_openmp])

data_seq = np.array([list(map(int, row.split())) for row in data_seq])
```

```python
sizes = np.array([5, 10, 25, 50, 100, 250, 500, 750, 1000, 1100])

fig, ax = plt.subplots()
ax.plot(sizes, np.mean(data_openmp, axis = 1), label = 'Среднее, c openmp')
ax.plot(sizes, np.mean(data_seq, axis = 1), label = 'Среднее, без openmp')
ax.set_xlabel('Размер')
ax.set_ylabel('Время')
ax.set_title('Зависимость времени от размера')
ax.legend()
```
